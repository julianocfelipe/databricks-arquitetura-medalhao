# 006 — Destruindo o Ambiente
## Arquitetura Medalhão · Loja Virtual

Remove **todos** os schemas da arquitetura medalhão para começar do zero.

| Schema | O que é apagado |
|--------|-----------------|
| `bronze` | Tabelas Delta da camada Bronze |
| `silver` | Tabelas Delta da camada Silver |
| `gold`   | Tabelas dimensionais da camada Gold |
| `landing` | Volume `dados` (CSV brutos) + schema |

## 1. Remover todos os schemas (CASCADE)

```python
%sql
-- Apagar todas as tabelas da camada bronze
DROP SCHEMA IF EXISTS workspace.bronze CASCADE;

-- Apagar todas as tabelas da camada silver
DROP SCHEMA IF EXISTS workspace.silver CASCADE;

-- Apagar todas as tabelas da camada gold
DROP SCHEMA IF EXISTS workspace.gold CASCADE;

-- Apagar todas as tabelas e volumes da camada landing
DROP SCHEMA IF EXISTS workspace.landing CASCADE;
```

## 2. Confirmar se tudo foi limpo

```python
%sql
SHOW SCHEMAS IN workspace;
```

## ✅ Ambiente destruído!

Os schemas `landing` · `bronze` · `silver` · `gold` foram removidos.
