# 02 — Landing → Bronze

Copia as tabelas da camada **Landing** para a camada **Bronze** em formato **Delta Lake**.

```python
LANDING_SCHEMA = "workspace.landing"
BRONZE_SCHEMA = "workspace.bronze"

TABLES = [
    "clientes",
    "produtos",
    "pedidos"
]
```

```python
spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {BRONZE_SCHEMA}
""")
```

```python
results_bronze = []

for table in TABLES:
    print(f"Processando: {table}")

    df = spark.table(f"{LANDING_SCHEMA}.{table}")

    (
        df.write
            .format("delta")
            .mode("overwrite")
            .saveAsTable(f"{BRONZE_SCHEMA}.{table}")
    )

    count = df.count()

    results_bronze.append({
        "table": table,
        "rows": count
    })

    print(f"OK: {count} registros")
```
