# 05 — Reset

Remove as tabelas de todas as camadas e limpa o Volume da Landing para recomecar do zero.

```python
base_tables = ["clientes", "produtos", "pedidos"]
gold_tables = ["dim_cliente", "dim_produto", "fato_pedido"]

for schema in ["workspace.landing", "workspace.bronze", "workspace.silver"]:
    for table in base_tables:
        spark.sql(f"DROP TABLE IF EXISTS {schema}.{table}")
        print(f"DROP TABLE {schema}.{table}")

for table in gold_tables:
    spark.sql(f"DROP TABLE IF EXISTS workspace.gold.{table}")
    print(f"DROP TABLE workspace.gold.{table}")
```

```python
dbutils.fs.rm(
    "/Volumes/workspace/landing/dados",
    recurse=True
)
```
